## Sortieren des DataFrames

```python
import pandas

df = pandas.read_csv('datas/sales.csv')

#Index setzen
df = df.set_index('Order ID')

# In Datetime umwandeln
df['Order Date'] = pandas.to_datetime(df['Order Date'])
df['Ship Date'] = pandas.to_datetime(df['Ship Date'])

# Sortieren
df = df.sort_values('Order Date')
df
```

## Werte filtern

```python
import pandas

df = pandas.read_csv('datas/sales.csv')

#Index setzen
df = df.set_index('Order ID')

# In Datetime umwandeln
df['Order Date'] = pandas.to_datetime(df['Order Date'])
df['Ship Date'] = pandas.to_datetime(df['Ship Date'])

# Sortieren
df = df.sort_values('Order Date')

# Filtern
df = df.query('"2012-01-01" <= `Order Date` & `Order Date` <= "2012-12-31"')
df = df.query('Region == "Europe"')
df
```

## Filtern und sortieren

```python
df = df.query('Country == "Germany" or Country == "France"').sort_values('Country')
df
```

## Neue Werte reinschreiben

```python
df['Fulfilment Delta'] = df['Ship Date'] - df['Order Date']
df
```
